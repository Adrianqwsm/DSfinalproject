```python
import numpy as np
import matplotlib.pyplot as bar
import pandas as pd
import matplotlib.pyplot as bar
```

```python
###I have found out that of the 198 countries, these are the amount of counties that speak the languages###
data = pd.DataFrame({'English': [78],
                     'Mandarin': [6],
                     'Arabic': [26],
                     'French': [34],
                     'Spanish': [23]})
display(data)
```

```python
###Here I'm just trying to make the information more readable.
total_English = data.English.sum()
print('The total number of English speaking countries is: %s' % total_English)

total_Mandarin = data.Mandarin.sum()
print('The total number of Mandarin speaking countries is: %s' % total_Mandarin)

total_Arabic = data.Arabic.sum()
print('The total number of Arabic speaking countries is: %s' % total_Arabic)

total_French = data.French.sum()
print('The total number of French speaking countries is: %s' % total_French)

total_Spanish = data.Spanish.sum()
print('The total number of Spanish speaking counties is: %s' % total_Spanish)
```

```python
### Here I'm using a really basic formula to calculate the percentages, then I will round it off to the closest whole number ###
def is_what_percent_of(num_a, num_b):
    return (num_a / num_b) * 100


print(is_what_percent_of(78, 198))  ### For English
print(is_what_percent_of(6, 198))  ### For Mandarin
print(is_what_percent_of(26, 198)) ### For Arabic
print(is_what_percent_of(34, 198))  ### For French
print(is_what_percent_of(23, 198)) ### For Spanish

### I use (num_b'(198)) because there are 198 entries (countries) in the file.
```

```python
###Percentages now placed into the visualization bar graph.
left = [1, 2, 3, 4, 5]
height = [39, 3, 13, 17, 12]
tick_label = ['English', 'Mandarin', 'Arabic', 'French', 'Spanish']
plt.bar (left, height, tick_label = tick_label,
         width = 0.8, color = ['blue', 'yellow'])
bar.xlabel('Common Languages')
bar.ylabel('Percentage of countries (total of 198)')
bar.title('Popularity of Common Languages')
plt.show()
```
